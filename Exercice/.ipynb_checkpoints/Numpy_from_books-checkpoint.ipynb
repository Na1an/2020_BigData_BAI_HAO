```python
#布尔型索引
names = np.array(['alice','bob','celina','diamond','elise','bob','tim'])
names
```

```python
data = np.random.randn(7,4)
data
```

```python
names == 'bob'
```

```python
data[names=='bob']
```

```python
data[~(names=='bob'),:1]
```

```python
mask = (names == 'bob') | (names == 'tim')
mask
```

```python
data[~mask]
```

```python
data[data < 0] = 0
data
```

```python
data[names != 'bob'] = 7
data
```
